```python
# –§–æ–∫—É—Å–∏—Ä—É–µ–º—Å—è –Ω–∞ –ê—Ä–≥–µ–Ω—Ç–∏–Ω–µ
argentina_matches = matches_df[
    (matches_df['home_team.home_team_name'] == 'Argentina') | 
    (matches_df['away_team.away_team_name'] == 'Argentina')
].copy()

print(f"–ê—Ä–≥–µ–Ω—Ç–∏–Ω–∞ —Å—ã–≥—Ä–∞–ª–∞ {len(argentina_matches)} –º–∞—Ç—á–µ–π –Ω–∞ –ß–ú-2022")
print("=" * 50)

# –°—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –ê—Ä–≥–µ–Ω—Ç–∏–Ω—ã
argentina_goals_for = 0
argentina_goals_against = 0
wins = 0

for _, match in argentina_matches.iterrows():
    if match['home_team.home_team_name'] == 'Argentina':
        goals_for = match['home_score']
        goals_against = match['away_score']
    else:
        goals_for = match['away_score'] 
        goals_against = match['home_score']
    
    argentina_goals_for += goals_for
    argentina_goals_against += goals_against
    if goals_for > goals_against:
        wins += 1

print(f"–ü–æ–±–µ–¥—ã: {wins}/{len(argentina_matches)}")
print(f"–ó–∞–±–∏—Ç–æ –≥–æ–ª–æ–≤: {argentina_goals_for}")
print(f"–ü—Ä–æ–ø—É—â–µ–Ω–æ –≥–æ–ª–æ–≤: {argentina_goals_against}")
print(f"–†–∞–∑–Ω–∏—Ü–∞ –º—è—á–µ–π: +{argentina_goals_for - argentina_goals_against}")
```

```python
# –°–æ–∑–¥–∞—ë–º –ø—Ä–æ—Å—Ç–æ–π –∏ –ø–æ–Ω—è—Ç–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫
labels = ['–ó–∞–±–∏—Ç–æ', '–ü—Ä–æ–ø—É—â–µ–Ω–æ']
values = [argentina_goals_for, argentina_goals_against]
colors = ['#6CACE4', '#FF6B35']  # –°–∏–Ω–∏–π –∏ –æ—Ä–∞–Ω–∂–µ–≤—ã–π - —Ü–≤–µ—Ç–∞ –ê—Ä–≥–µ–Ω—Ç–∏–Ω—ã

plt.figure(figsize=(8, 6))
bars = plt.bar(labels, values, color=colors, alpha=0.8)
plt.title('–ê—Ä–≥–µ–Ω—Ç–∏–Ω–∞ –Ω–∞ –ß–ú-2022: –ó–∞–±–∏—Ç—ã–µ –∏ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –≥–æ–ª—ã', fontsize=14, fontweight='bold')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–æ–ª–æ–≤', fontsize=12)

# –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, 
             f'{value}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
